Assume without loss of generality that the optimization problem is to ﬁnd the **minimum score** (OPT)

Find all Bayesian networks $G$ that are within a factor $\rho$ of optimal (***credible network, Def. 1.***), i.e.,
  $$OPT \leq \sigma(G) \leq \rho \cdot OPT$$
For some $\rho \geq 1$, equivalently,
  $$OPT \leq \sigma(G) \leq OPT+\epsilon$$
where $\epsilon = (\rho-1)\cdot OPT$. And we call this problem $\epsilon$-Bayesian Network Structual Learning ($\epsilon$BNSL)

For BIC/MDL and BDeu, a good choice of $\rho$ is related to **Bayes factor**.

Two advantages
* Only consider *credible* models
* Efficient and scale to Bayesian networks with almost 60 variables

## Scoring Functions $\sigma$
Important property: **decomposability**, i.e., $\sigma(G)=\sum_i \sigma(V_i,\Pi_i)$

* BIC <br/>
  $$\sigma(G)=max_\theta L_{G,I}(\theta)-t(G)w$$
  where $L_{G,I}(\theta)$ is the log likelyhood, $w=\frac{log(N)}{2}$ and $t(G)=|G|=number of parameters in G$
* BDeu <br/>
  $$\sigma(G)=\sum_{i=1}^{n} \sum_{j=1}^{r_{\Pi_i}} log \frac{\Gamma(\alpha)}{\Gamma(\alpha+n_{ij})} + \sum_{i=1}^{n} \sum_{j=1}^{r_{\Pi_i}} \sum_{k=1}^{r_i} log \frac{\Gamma(\frac{\alpha}{r_i}+n_{ijk})}{\Gamma(\frac{\alpha}{r_i})}$$

## Bayes Factor
* Model selection criterion
* Alternative to significance test

$$BF(G_0,G_1)=\frac{P(I|G_0)}{P(I|G_1)}$$

BF can be obtained using $P(G|I)$ or $P(G,I)$

$$\frac{P(G_0|I)}{P(G_1|I)}
    =\frac{\frac{P(I|G_0)\cdot P(G_0)}{P(I)}}{\frac{P(I|G_1)\cdot P(G_1)}{P(I)}}
    =\frac{P(G_0,I)}{P(G_1,I)}$$
    
Typical to assume $P(G_0)=P(G_1)$, so,
  $$BF(G_0,G_1)=\frac{P(I|G_0)}{P(I|G_1)}=\frac{P(G_0|I)}{P(G_1|I)}=\frac{P(G_0,I)}{P(G_1,I)}$$

---

BIC, BDeu and BF
* $BIC(I,G_1)-BIC(I,G_0)=log BF(G_0,G_1)$
* $BDeu(I,G_1)-BDeu(I,G_0)=log BF(G_0,G_1)$

---

scale of $BF(G_0,G_1)$ : Strength of evidence
* 1-3 : anecdotal evidence
* 3-20 : positive evidence that $G_0$ is better
* 20-150 : strong evidence in favor of $G_0$
* 150-$\infty$ : very strong evidence

The $\epsilon$BNSL problem using BIC or BDeu given a desired BF can be written as,
$$OPT \leq \sigma(G) \leq OPT+logBF$$
